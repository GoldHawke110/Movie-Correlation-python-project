```python
## Library imports 

import pandas as pd 
import seaborn as sns
import numpy as np 

import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) # This will adjust the configuration of the plots
pd.options.mode.chained_assignment = None



# Data read 

df = pd.read_csv(r'D:\Users\Luke\Desktop/movies.csv')
```

```python
## Data visualisation 

df.head()
```

```python
## missing data 

for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Column data types 

df.dtypes
```

```python
# Changing data types 


df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# creating corrected year column

df = df.dropna()
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
##df.head()
```

```python
# Removing restrictions (displaying ALL rows)

## pd.set_option('display.max_rows', None)
```

```python
# dropping duplicates 

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Higher the budget = higher the income (Correlation between budget and gross income)
```

```python
# scatter plot: budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
## making the graph title 
plt.title('budget vs gross earnings')
## naming X and Y axis
plt.xlabel('gross earnings')
plt.ylabel('Film budget')

plt.show
```

```python
# plot using seaborn 

sns.regplot(x='budget', y ='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# using seaborn heatmap

correlation_matrix = df.corr(method= 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
